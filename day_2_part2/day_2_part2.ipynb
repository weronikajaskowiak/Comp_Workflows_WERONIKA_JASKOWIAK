Explain the quality control steps. 

Read quality assessment: checks the quality of the raw sequencing reads (FastQC),
1) Trimming and filtering: low-quality bases and adapter sequences are removed from the reads (Trim Galore!),
rRNA content: measure the percentage of reads that align to ribosomal RNA (rRNA) (high levels are undesirable),
2) Duplication level: percentage of duplicated reads, which can indicate issues with library diversity or over-amplification during PCR,
3) GC content contribution: check if the GC content of the reads matches the expected GC content of the organismâ€™s genome,
4) Alignment quality: determine how well the reads map to the reference genome or transcriptome (STAR/Salmon), 
5) Gene expression quantification: measure the number of reads mapped to each gene or transcript.
6) MultiQC summary report: aggregate the results from different QC tools into a single comprehensive report.

Please give additional information on : 
1) Ribosomal rRNA: RNA-seq typically aims to capture messenger RNA (mRNA), so high levels of rRNA contamination (above 1-5%) indicate poor quality. If a high percentage of rRNA is found in the reads, it suggests that the library preparation wasn't done properly. Only two samples, SNI_Sal_4 and SNI_Sal_2, have a very high percentage of rRNA (>20%).
2) Duplication: generally, a higher percentage of duplicated reads is welcome, as it indicates better sequencing depth and lower contamination from sequencing errors. The overall ratio of duplicates to unique reads suggests that these libraries have been fully sequenced, and further sequencing would likely result in only a small number of new unique reads.
3) GC Content: the GC content of reads should correspond to the expected GC content of the studied genome. The expected GC content should range between 40% and 45%. It can be observed that 14 of the 32 samples have two peaks, indicating contamination or technical issues. Among those samples is SNI_Sal_2, which also has a high rRNA content.
